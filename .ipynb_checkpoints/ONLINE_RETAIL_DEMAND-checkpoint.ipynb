# INVENTORY STOCK IMPROVEMENT
## Purpose of the project
The purpose of this project is to predict the amount of a product an Online retailer will need to have in stock to supply its customers from diferent countries. 

## Implications
The results from this project can help managers to organize and maintain their inventory and save money. 

## Data origin and description
The data is from the UCI Machine Learning Repository, can be accessed in my github (https://github.com/jpinzonc/di_10_17_retail_price_pred) or at the original site (https://archive.ics.uci.edu/ml/machine-learning-databases/00352/). 

The description of the data indicates that the file, in excel format, contains purchase transactions from 01/12/2010 to 09/12/2011 from a UK-based online store that sales gifts and other small items. The customers, mostly wholesalers, are locted in the UK as well as in other countries. 

The file contains eithg columns with the follwoing information:
- InvoiceNo: Invoice number. Categorical. A unique 6-digit number assigned to each transaction. If this code starts with letter 'c', it indicates a cancellation. 
- StockCode: Product or item code. Categorical. A unique 5-digit number assigned to each distinct product. 
- Description: Product name. Text. 
- Quantity: The quantities of each product per transaction. Numeric.	
- InvoiceDate: Invoice Date and time. Numeric, the day and time when each transaction was generated. 
- UnitPrice: Unit price. Numeric, Product price per unit in sterling. 
- CustomerID: Customer number. Categorical. A unique 5-digit number assigned to each customer. 
- Country: Country name. Categorical, the name of the country where each customer resides.


### Data Exploration and Visualization
Reading the data from the excel file and explore, describe and visualizate it.  

```python
# Importing the necesaty libraries:
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Reading the file into Python
xlfile = 'Online Retail.xlsx'
retail = pd.ExcelFile(xlfile)
```

```python
# Extracting the spreadsheet with the data and creating a data frame
df_retail = retail.parse("Online Retail")
```

```python
display(df_retail.head())
display(df_retail.describe())
df_retail.shape
```

There are some negative values in both features of interest, Quantity and UnitPrice. 

```python
# Transform customer ID to string 
df_retail.CustomerID = df_retail.CustomerID.astype(str)
df_retail[df_retail.Quantity<=0].head()
```

These appear to be cancellations and discounts. 

The first step is to remove the rows that correspond to order cancellations. 

Additionally, there some orders, three of them, that start with an 'A'. Their description says 'adjust bad debt'. 

Some values in the Invoice column are numeric therefore I transform them to string for the script to be able to remove the rows with letters. 

Data with positive UnitPrice are maintained. 

```python
df_retail.InvoiceNo = df_retail.InvoiceNo.astype(str)
df_retail = df_retail[df_retail.InvoiceNo.str.contains('A|C', na=False)==False]
df_retail=df_retail[df_retail.UnitPrice>0]
df_retail.shape
```

The description of the products, also includes transactions such as:
Postage fees (POST), Manual entries - not clear what this mean (M or m), PADS (PADS), AMAZON FEES (AMAZON FEES), BANK CHARGES (BANK CHARGES) and SAMPLES (SAMPLES). 
This do not correspond to products sold by the company by payments it made. 

```python
df_retail=df_retail[df_retail.StockCode!='POST']
df_retail=df_retail[df_retail.StockCode!='M']
df_retail=df_retail[df_retail.StockCode!='PADS']
df_retail=df_retail[df_retail.StockCode!='S']
df_retail=df_retail[df_retail.StockCode!='AMAZONFEE']
df_retail=df_retail[df_retail.StockCode!='BANK CHARGES']
df_retail=df_retail[df_retail.StockCode!='m']
df_retail.shape
```

After checking the countries there us a category 'unspecified' that do not correspond to any country.  

```python
display(pd.DataFrame(df_retail.Country.unique(), columns=['Country']).sort_values(by='Country').reset_index(drop=True))

df_retail = df_retail[df_retail.Country.str.contains('Unspecified', na=False)==False]

df_retail.shape
```

```python
df_retail.describe()
```

### Visualization
Now We start with the visualization. Fist, look at the numeric variables, Unit Price and Quantity.

From the graph below, it seems there are a few outliers the Quantity. Values higher than 60000. 

```python
df = df_retail[['Quantity', 'UnitPrice']]
n_max = (max((df.Quantity.max(), df.UnitPrice.max())))
data = df.values

plt.rcParams["figure.figsize"] = (20,10)
plt.boxplot(data, 0, 'gD')

ticks = ['Quantity', 'Price']
plt.xticks(range(1, len(ticks)+1,1), ticks)
plt.xlabel('Feature')
plt.ylabel('Value')
plt.xlim(0, len(ticks)+1)
plt.ylim(0, n_max)
```

```python
df_retail[df_retail.Quantity>60000]
```

While the outliers appear to be real products. They seem to be something very small and only sold in the UK. 
Given they are only two transactions, they will be removed from the dataset

```python
df_retail=df_retail[df_retail.Quantity<60000]
df_retail.shape
```

```python
df = df_retail[['Quantity', 'UnitPrice']]
n_max = (max((df.Quantity.max(), df.UnitPrice.max())))
data = df.values

plt.rcParams["figure.figsize"] = (20,10)
plt.boxplot(data, 0, 'gD')

ticks = ['Quantity', 'Price']
plt.xticks(range(1, len(ticks)+1,1), ticks)
plt.xlabel('Feature')
plt.ylabel('Value')
plt.xlim(0, len(ticks)+1)
plt.ylim(0, n_max)
plt.title("Quantity and Price Boxplots")
plt.savefig('Quantity_and_UnitPrice.jpg')
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.scatter(df_retail.UnitPrice, df_retail.Quantity)
plt.xlabel("Unite Price")
plt.ylabel("Quantity")
```

There appears to be no correlation with these two features, atleast not linear one. This is not considering each product separately. 

Next we examine the relation between product and countries. 

```python
df_country_prod = df_retail.groupby(['Country'])['StockCode'].count()
df_country_prod = df_country_prod.reset_index(drop=False)
df_country_prod
df_country_prod.sort_values(by='StockCode')
```

```python
df_country_prod1=df_country_prod[(df_country_prod.StockCode>1000) & (df_country_prod.StockCode<10000)]
df_country_prod2=df_country_prod[(df_country_prod.StockCode>100) & (df_country_prod.StockCode<1000)]
df_country_prod3=df_country_prod[(df_country_prod.StockCode<100)]

objects1 = df_country_prod1.Country
pos1 = np.arange(len(objects1))
performance1 = (df_country_prod1.StockCode)

objects2 = df_country_prod2.Country
pos2 = np.arange(len(objects2))
performance2 = (df_country_prod2.StockCode)

objects3 = df_country_prod3.Country
pos3 = np.arange(len(objects3))
performance3 = (df_country_prod3.StockCode)

plt.rcParams["figure.figsize"] = (20,15)
f, ((ax1, ax2, ax3)) = plt.subplots(3, 1, sharey=False)
f.suptitle('AMOUNT OF PRODUCTS ORDERED BY COUNTRY FROM THE RETAIL STORE \n Excluding UK ', fontsize=20)
f.text(0.5, 0.04, 'COUNTRY', ha='center',fontsize=30)
f.text(0.04, 0.5, 'TOTAL AMOUNT OF PRODUCTS ORDERED PER COUNTRY', va='center', rotation='vertical', fontsize=25)

ax1.bar(pos1, performance1, align='center', alpha=0.5, color='b') 
ax1.set_title("Amounts larger than 1000")
ax1.set_xticks(pos1)
ax1.set_xticklabels(objects1, fontsize=11)

ax2.bar(pos2, performance2, align='center', alpha=0.5, color='c') 
ax2.set_title("Amounts between 100 and 1000")
ax2.set_xticks(pos2)
ax2.set_xticklabels(objects2, fontsize=11)

ax3.bar(pos3, performance3, align='center', alpha=0.5, color ='g') 
ax3.set_title("Amounts smaller than 100")
ax3.set_xticks(pos3)
ax3.set_xticklabels(objects3, fontsize=11)


plt.savefig('Products_amt_per_Country.jpg')
```
